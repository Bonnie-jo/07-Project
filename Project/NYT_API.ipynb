```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time


# Import API key
from pprint import pprint
from config import api_key_nyt 
from pandas.io.json import json_normalize

import urllib3
urllib3.disable_warnings()
```

```python
keyword_mod = input('What keyword are you looking for? eg. Man, Woman, Boy, Girl, Children, Couple ')
```

```python
state = input('What State are you looking for? eg. Florida, California ')
```

```python
#starting URL

keywords = '"' + state + '%20' + keyword_mod +'"'

url = 'https://api.nytimes.com/svc/search/v2/articlesearch.json?q=' + keywords + '&api-key=' + api_key_nyt
#"https://www.accunewsapp.com/api/v1/articles/?lon=28.9025&lat=82.5926&apiKey=" + api_key + "&keyword=" + keywords + "%5C&radius=300"

print(url)
```

```python
titles_1 = []
summary_1 = []
link_1 = []
time_1 = []
section_1 = []
subsection_1 = []
articles_list = []
category = []

articles = requests.get(url,verify=False).json()
articles_list = [article for article in articles["response"]["docs"]]

for page in range(0, 100):
   # create query with page number
   query_url = f"{url}&page={str(page)}"
   articles = requests.get(query_url).json()

   # Add a seven second interval between queries to stay within API query limits
   time.sleep(7)
   # loop through the response and append each article to the list
   for article in articles["response"]["docs"]:
    articles_list.append(article)
    titles_1.append(article['headline']['main'])
    summary_1.append(article['snippet'])
    time_1.append(article['pub_date'])
    link_1.append(article['web_url'])
    category.append(state + ' ' + keyword_mod)
    try: section_1.append(article['section_name'])
    except: section_1.append('n/a')
    try: subsection_1.append(article['section_name'])
    except: subsection_1.append('n/a')

#articles_list
#print(articles['next'])
#i = 0
#while True:

#for result in articles['response']['docs']:
    #titles_1.append(result['headline']['main'])
    #summary_1.append(result['snippet'])
    #time_1.append(result['pub_date'])
    #link_1.append(result['web_url'])
    #section_1.append(result['section_name'])
    #time.sleep(6.01)
    
    #i = i + 1
    #if i >1000: 
        #break
    #articles = requests.get(articles['next'],verify=False).json()
        
    #if articles == {}:
        #break
#pprint(articles)
```

```python
nyt_df_dict = {'Title': titles_1,
        'Summary': summary_1,
        'Link': link_1,
        'Time': time_1,
        'Section': section_1,
        'Subsection': subsection_1
        'Category' : category}

nyt_df = pd.DataFrame(nyt_df_dict)

nyt_df.to_csv("nyt_" + state + "_" + keyword_mod + "_data.csv", index=False)
nyt_df
```
