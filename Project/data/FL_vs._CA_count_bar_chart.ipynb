```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Read CSV file
nyt_man_csv_path = "filtered_nyt_California_Man_data.csv"
nyt_CA_Man_df = pd.read_csv(nyt_man_csv_path)

# First index of tuple returned by shape contains the number of index/row in dataframe
NYT_CA_Man_num = nyt_CA_Man_df.shape[0]
 
print('Number of Rows in NYT CA Man dataframe : ' , NYT_CA_Man_num)
```

```python
# Read CSV file
nyt_FL_man_csv_path = "filtered_nyt_Florida_Man_data.csv"
nyt_FL_Man_df = pd.read_csv(nyt_FL_man_csv_path)

# First index of tuple returned by shape contains the number of index/row in dataframe
NYT_FL_Man_num = nyt_FL_Man_df.shape[0]
 
print('Number of Rows in NYT FL Man dataframe : ' , NYT_FL_Man_num)
```

```python
# Read CSV file
Loc_FL_man_csv_path = "filtered_Local_Florida Man.csv"
Loc_FL_Man_df = pd.read_csv(Loc_FL_man_csv_path)

# First index of tuple returned by shape contains the number of index/row in dataframe
Loc_FL_Man_num = Loc_FL_Man_df.shape[0]
 
print('Number of Rows in Local FL Man dataframe : ' , Loc_FL_Man_num)
```

```python
labels = ['CA', 'FL']
CA_count = [5, 110]
FL_count = [61, 148]

x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, CA_count, width, label='Local')
rects2 = ax.bar(x + width/2, FL_count, width, label='National')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Number of Articles Published RE: "State" Man ')
ax.set_title('Article Count Comparison')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()


def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')


autolabel(rects1)
autolabel(rects2)

fig.tight_layout()

plt.show()
```
