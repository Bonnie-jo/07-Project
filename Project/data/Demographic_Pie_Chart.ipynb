```python
#Build Local Pie Chart

#import csvs
local_man_to_load = "Local_Florida Man.csv"
local_woman_to_load = "Local_Florida Woman.csv"
local_boy_to_load = "Local_Florida Boy.csv"
local_girl_to_load = "Local_Florida Girl.csv"
local_couple_to_load = "Local_Florida Couple.csv"

#read csv
local_man_data = pd.read_csv(local_man_to_load)
local_woman_data = pd.read_csv(local_woman_to_load)
local_boy_data = pd.read_csv(local_boy_to_load)
local_girl_data = pd.read_csv(local_girl_to_load)
local_couple_data = pd.read_csv(local_couple_to_load)

#combine all dataframes
df_all_local = pd.concat([local_man_data, local_woman_data, local_boy_data, local_girl_data, local_couple_data])
df_all_local
```

```python
#Group by Gender
local_grouped_gender = df_all_local.groupby(['Category'])
local_grouped_counts = local_grouped_gender['Category'].count()
local_gender_df = pd.DataFrame({'Demographic': local_grouped_counts})
local_gender_df
```

```python
colors = ["darkorange","royalblue","green","crimson","mediumpurple"]
explode = (0, 0, 0, 0.1, 0)
local_grouped_counts.plot(kind = "pie",
                y=local_gender_df.keys(),
                figsize=(20,10),
                fontsize=20,
                labels = None,
                explode=explode,
                colors=colors,
                autopct="%1.1f%%",
                startangle=140,
                title ="Stories from Local News about Florida by Demographic",
                shadow=True
                )
plt.legend(labels=local_gender_df.index, bbox_to_anchor=(1,0.5), loc="center right")
#plt.rcParams.update()
#plt.rcParams.update({'font.size': 22})
plt.axis('off')
plt.axis('equal')
#plt.tight_layout()

plt.savefig('Local Florida Demographics.png')
```
